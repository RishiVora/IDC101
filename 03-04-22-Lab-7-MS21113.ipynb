```python
from math import ceil, floor

def trunc(x, place=6) :
  truncated = int(x * (10**place))/10**place
  return truncated

def round_up(x, place=6) :
  rounded_up = ceil(x * (10**place))/10**place
  return rounded_up

def round_down(x, place=6) :
  rounded_down = floor(x * (10**place))/10**place
  return rounded_down

def T(x) :
  powers = [x]
  for n in range(1,4) :
    powers.append(x**n)
  y = trunc(2*powers[2]) - trunc(5.2*powers[1]) + trunc(4.2*powers[0]) + 2
  return y

def rUP(x) :
  powers = [x]
  for n in range(1,4) :
    powers.append(x**n)
  y = round_up(2*powers[2]) - round_up(5.2*powers[1]) + round_up(4.2*powers[0]) + 2
  return y

def rDN(x) :
  powers = [x]
  for n in range(1,4) :
    powers.append(x**n)
  y = round_down(2*powers[2]) - round_down(5.2*powers[1]) + round_down(4.2*powers[0]) + 2
  return y

vals = [4.1625, 2.1672]

for x in vals :
  print(f'x = {x}')
  print(f'T = {T(x)}, rUP = {rUP(x)}, rDN = {rDN(x)}')
  print(f'err_up = {abs((T(x)-rUP(x))/T(x))}, err_dn = {abs((T(x)-rDN(x))/T(x))}\n')
```

```python
from random import choice

times = int(input('How many times you want to repeat? : '))
l = [-1,0,1]
final = []

for r in range(0,times) :
  x = sum([choice(l) for n in range(0,100)])
  final.append(x)

print('Average =', sum(final)/times)
```
