```python
age = int(input('Enter your age: '))

if age >= 0 and age <= 12 : print('The person is kid')
if age >= 13 and age <= 17 : print('The person is teen')
if age >= 18 and age <= 59 : print('The person is adult')
if age >= 60 : print('The person is senior citizen')
```
