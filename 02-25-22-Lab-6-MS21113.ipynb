```python
import string

def crypt(s) :
  result = ''
  for char in s :
    char = char.upper()
    if char == 'Y' : result += 'a'
    elif char == 'Z' : result += 'b'
    else : result += chr(ord(char)+2).lower()
  print(result)

def decrypt(s) :
  result = ''
  for char in s :
    char = char.upper()
    if char == 'A' : result += 'y'
    elif char == 'B' : result += 'z'
    else : result += chr(ord(char)-2).lower()
  print(result)

s = input('Enter a string: ').upper()
crypt(s)
decrypt(s)
```

```python
from random import random as rnd

n = int(input('Enter the number of points to be generated: '))

def pi(n) :
  x,y = [],[]
  cir = 0

  for i in range(n) :
    x.append(rnd())    #x.append(rnd.uniform(-1,1))
    y.append(rnd())    #y.append(rnd.uniform(-1,1))

  for xc,yc in zip(x,y) :
    if ((xc-0.5)**2)+((yc-0.5)**2) <= 0.25 :
      cir += 1
  
  value = 4*(cir/n)
  return value

pi(n)
```
