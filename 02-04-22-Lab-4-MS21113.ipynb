```python
num = int(input('Enter a number: '))
for i in range(2, num) :
  if num % i == 0 :
    print(f'{num} is not prime')
    break
  else :
    print(f'{num} is prime')
    break
```

```python
num, up = False, False

while True :
  string = input('Enter a string: ')
  if len(string) < 8 :
    print('Invalid String')
    continue

  for i in string :
    if i.isdigit() :
      num = True
      break

  for i in string :
    if i.isupper() :
      up = True
      break

  if num is True and up is True:
    print('Valid String')
    break
  else :
    print('Invalid String')
```

```python
num = int(input('Enter a number: '))

primes = []
msg = ''

for i in range(2, num+1) :
  for j in range(2, i):
    if (i % j) == 0:
      break
  else:
    primes.append(i)

for prime in primes :
  pow = 0
  while True:
    if num % prime == 0 :
      pow += 1
      num = num // prime
      if num == 1 : break
    else : break
  if pow == 0 : continue
  msg += f'({prime}^{pow}) * '
print(msg[:-2])
```

```python
def checkprime(num) :
  if num <= 0 : return 'Invalid input, enter positive number'
  elif num == 1 : return '1 is a special number'
  else : pass
  for i in range(2, num):
    if (num % i) == 0:
      return f'{num} is not prime'
  else:
    return f'{num} is prime'

n = int(input('Enter a number: '))
checkprime(n)
```
