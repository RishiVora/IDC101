````python
#Q5-a
x = int(input('Enter a number: '))
if x%2 == 0 :
  print(f'{x} is even')
else :
  print(f'{x} is odd')
````

````python
#Q5-b
total = 0
for i in range(101,299) :
  if i%2 == 0 :
    total += i
print('Sum is', total)
````

````python
#Q5-c
x = int(input('Enter a number: ')); j = 1

for i in range(2,x) :
  if x%i != 0 : continue
  j = 0

if j == 1 : print(x, 'is prime')
if j == 0 : print(x, 'is not prime')
````

````python
#Q5-d
sum = 0
for i in range(2,200):
  for j in range(2,i):
    if i % j == 0:
      break
  else:
    sum += i
print(sum)
````

````python
#Q5-e (for loop)
x = 7
f = 1
for i in range(1,x+1):
  f *= i
print(f)

#Q5-e (while loop)
x = 7; f = 1; i = 1
while i<=x :
  f *= i
  i += 1
print(f)
````
