```python
# run this once

from math import sin, cos
import random
import numpy as np
import matplotlib.pyplot as plt 

def f1(x) :
  y = x**4 + 3*x**2 - 2*x - 10
  return y

def df1(x) :
  y = 4*x**3 + 6*x - 2
  return y

def f2(x) :
  try : y = np.array([(i - sin(i) - 5) for i in x])
  except : y = x - sin(x) - 5
  return y

def df2(x) :
  try : y = np.array([(1 - cos(i)) for i in x])
  except : y = 1 - cos(x)
  return y

def showaxis() :
  for p in [axs[0, 0], axs[0, 1], axs[1, 0], axs[1, 1]] :
    p.plot(np.arange(-100,100,1),[0]*len(np.arange(-100,100,1)), color='black')
    p.plot([0]*len(np.arange(-100,100,1)),np.arange(-100,100,1), color='black')
```

```python
#Bisection Method

def bisection(fn) :

  a = float(input("Enter a such that f(a) > 0 : "))
  b = float(input("Enter b such that f(b) < 0 : "))

  if fn(a) < 0 or fn(b) > 0 :
    print('Invalid values')
    return

  i = 0

  while True :
    c = (a+b)/2

    if -0.001 < fn(c) < 0.001 :
      print('Root is', c)
      print(f'Loop was repeated {i} times\n')
      return
    
    if fn(a)*fn(b) < 0 :
      c = random.uniform(a,b)
      if fn(c) > 0 : a = c
      elif fn(c) < 0 : b = c
    
    i += 1

bisection(f1)
bisection(f2)
```

```python
#Newton-Raphson Method

def raphson(fn, dfn) :
  
  x = float(input('Enter a x-coordinate : '))
  i = 0

  while True :
    x = x - (fn(x)/dfn(x))
    if -0.001 < fn(x) < 0.001 :
      break
    i += 1
  
  print('Root is', x)
  print(f'Loop was repeated {i} times\n')
  return

raphson(f1, df1)
raphson(f2, df2)
```
